```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
fileFormat = ".pdf"
figsize = (4,4)

func = "ReLU"
plt.figure(figsize=figsize)
plt.plot([-100,0,100],[0,0,100], c="tab:green", linewidth=4, label=func)
plt.xlabel('x (Input)') 
plt.ylabel('f(x) (Output)')
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.legend()
plt.grid()
plt.savefig(func+fileFormat, bbox_inches='tight')
plt.show()
```

```python
func = "Sigmoid"
plt.figure(figsize=figsize)
plt.plot(np.linspace(-5,5,100),1/(1+(np.e**-np.linspace(-5,5,100))), c="tab:green", linewidth=4, label=func)
plt.xlabel('x (Input)') 
plt.ylabel('f(x) (Output)')
plt.xlim([-5,5])
plt.ylim([-1,1])
plt.legend()
plt.grid()
plt.savefig(func+fileFormat, bbox_inches='tight')
plt.show()
```

```python
func = "tanh"
plt.figure(figsize=figsize)
plt.plot(np.linspace(-5,5,100),np.tanh(np.linspace(-5,5,100)), c="tab:green", linewidth=4, label=func)
plt.xlabel('x (Input)') 
plt.ylabel('f(x) (Output)')
plt.xlim([-5,5])
plt.ylim([-1,1])
plt.legend(loc=2)
plt.grid()
plt.savefig(func+fileFormat, bbox_inches='tight')
plt.show()
```
